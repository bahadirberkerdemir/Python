```python
import matplotlib.pyplot as plt
import numpy as np # Sayısal işlemler için numpy kütüphanesini ekliyoruz

urunler = ["Elma", "Armut", "Muz", "Vişne"]
alis_fiyatlari = [15, 18, 22, 30]
satis_fiyatlari = [20, 25, 30, 40]

# X ekseni için ürün sayısı kadar sayı üretiyoruz [0, 1, 2, 3]
x = np.arange(len(urunler)) 
genislik = 0.35 # Sütunların genişliği

# Grafiği oluşturma
fig, ax = plt.subplots()

# Sütunları x konumundan biraz sola ve biraz sağa kaydırarak yan yana getiriyoruz
sutun1 = ax.bar(x - genislik/2, alis_fiyatlari, genislik, label='Alış', color='orange')
sutun2 = ax.bar(x + genislik/2, satis_fiyatlari, genislik, label='Satış', color='green')

# Başlıklar ve Etiketler
ax.set_title('Alış ve Satış Fiyat Karşılaştırması')
ax.set_ylabel('Fiyat (TL)')
ax.set_xticks(x) # X eksenindeki rakamların yerini belirle
ax.set_xticklabels(urunler) # Rakamların yerine ürün isimlerini yaz
ax.legend() # Hangi rengin ne olduğunu gösteren kutucuk (Alış/Satış)

plt.show()
```
