```python
import pandas as pd
# 1. Veriyi Sözlük olarak hazırla
veri = {
    "Ogrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep"],
    "Matematik": [85, 90, 78, 92],
    "Fizik": [70, 95, 80, 88]
}

# 2. Bu sözlüğü bir PANDAS TABLOSUNA dönüştür
df = pd.DataFrame(veri)

# 3. Tabloyu ekrana bas
print(df)

# 4. Sadece Matematik sütununu seç ve ortalamasını (mean) al
ortalama_mat = df["Matematik"].mean()

print("\nMatematik Ortalaması:", ortalama_mat)

# 5. Fizik Ortalaması ve Max değerini bul
ortalama_fiz = df["Fizik"].mean()

maks_fiz = df["Fizik"].max()

print(f"\nSinifin fizik ortalamasi {ortalama_fiz} iken sinifta fizikten alinan en yuksek not şudur: {maks_fiz} ...")

# 6. Fizik notu 80'den büyük olanları filtrele
caliskanlar = df[ df["Fizik"] > 80 ]

print("\n--- Fizikten 80 Üzeri Alanlar ---\n")
print(caliskanlar)

# 7. Matematik ve Fizik sütunlarını toplayıp 2'ye bölüyoruz
df["Genel_Ortalama"] = (df["Matematik"] + df["Fizik"]) / 2

print("\n--- Yeni Karne ---")
print(df)

# 8. 85 üstü alanları belirten tablo
takdir_belgesi = df[df["Genel_Ortalama"] >= 85]
print("\n\n --- Takdir Belgesine Hak Kazananlar ---")
print(takdir_belgesi)

# 9. Tabloyu 'takdir_listesi.xlsx' adıyla kaydediyoruz
# index=False komutu, baştaki 0,1,2,3 numaralarını Excel'e yazma demek (daha temiz görünür)
takdir_belgesi.to_excel("takdir_listesi.xlsx", index=False)

print("Dosya başarıyla Excel olarak kaydedildi!")

# 10. Var olan excel'i okut
yeni_veri = pd.read_excel("takdir_listesi.xlsx")

print(yeni_veri)
```
