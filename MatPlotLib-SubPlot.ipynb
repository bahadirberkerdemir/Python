```python
import matplotlib.pyplot as plt

uyg = ["WhatsApp", "Instagram", "Twitter", "Dier", "Uyku"]
zaman = [8, 2, 3, 4, 7]
renkler = ["green", "purple", "blue", "red", "yellow"]
patlama = [0.1, 0, 0, 0, 0]


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#-- Pasta Grafii Ayarlar

ax1.pie(zaman, labels = uyg, 
       autopct='%1.1f%%', 
       colors = renkler, 
       explode = patlama, 
       shadow=True, 
       startangle = 90)

ax1.axis('equal')
ax1.set_title("Gnlk Rutinim")

#-- Stun Grafii Ayarlar

genislik = 0.35

ax2.bar(uyg, zaman, genislik, color=renkler)
ax2.set_ylabel("Ayrilan Zaman")
ax2.set_xlabel("Aktiviteler")
ax2.set_title("AKTVTELERE AYRILAN ZAMANLAR")
ax2.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()   #Yazlar birbirine girmez

plt.show()


```
