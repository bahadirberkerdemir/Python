```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("buyuk_magaza_verisi.xlsx")

# 1) Sat fiyat orants var m bakmak iin "Scatter" tipi tablo

plt.figure(figsize=(10, 6))

# Scatter: x=Fiyat, y=Sat Adedi
# alpha=0.5 -> Noktalar %50 effaf yap ki st ste binenler koyu gzksn
plt.scatter(df["Fiyat"], df["Satis_Adedi"], color="purple", alpha=0.5)

plt.title("Fiyat ve Sat Adedi likisi")
plt.xlabel("rn Fiyat (TL)")
plt.ylabel("Satlan Adet")
plt.grid(True, linestyle='--', alpha=0.3)

plt.show()

# 2) Maazamz genel olarak ucuz mu pahal m rnler satyor bakmak iin "Scatter" tipi tablo


plt.figure(figsize=(10, 6))

# bins=20 -> Veriyi 20 eit paraya (kutucua) bl
# edgecolor -> Stunlarn kenarna izgi ek ki ayrlsnlar
plt.hist(df["Fiyat"], bins=20, color="skyblue", edgecolor="black")

plt.title("rn Fiyat Dalm (Histogram)")
plt.xlabel("Fiyat Aralklar")
plt.ylabel("Ka Tane rn Var?")

plt.show()
```
