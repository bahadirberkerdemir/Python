```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("buyuk_magaza_verisi.xlsx")


df["Ciro"] = df["Satis_Adedi"] * df["Fiyat"]       #Ciro hesaplanıp tabloya eklenir

Zengin_Satislar = df[df["Ciro"] >= 50000]           #50k üzeri ciro yapan ürünlerden oluşan yeni tablo
print(Zengin_Satislar)


Grouped_Zengin_Satislar = Zengin_Satislar.groupby("Kategori")["Ciro"].sum().sort_values(ascending = False)

menzil = len(Grouped_Zengin_Satislar)

pat = [0.1] + [0.0] * (menzil - 1)
renk = ["blue", "green", "yellow", "pink", "orange", "purple", "red"][:menzil]

plt.pie(Grouped_Zengin_Satislar, 
        labels = Grouped_Zengin_Satislar.index,
        colors = renk, 
        explode = pat,
        startangle = 90, 
        autopct='%1.1f%%', 
        shadow = True)


plt.title("Kategorilere Göre Ciro Dağilimi")
plt.show()
```
