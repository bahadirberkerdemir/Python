```python
import matplotlib.pyplot as plt
import pandas as pd

veri = {
    "Gun": ["Pzt", "Sal", "Çar", "Per", "Cum", "Cmt", "Paz"],
    "Ciro": [2000, 2400, 1800, 3000, 3500, 4200, 5000]
}

df = pd.DataFrame(veri)

df["Durum"] = "Normal"
df.loc[df["Ciro"] >= 3000, "Durum"] = "Yuksek"

print(df)

df.plot(x="Gun", y="Ciro", kind="bar", color = "Orange", legend = False)
plt.title("Haftalık Ciro Grafiği")
plt.ylabel("TL")
plt.show()


```
