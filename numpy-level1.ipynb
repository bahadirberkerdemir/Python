```python
import numpy as np
ham_puanlar = np.array([40, 50, 60, 45, 55])
son_puanlar = ham_puanlar + 5
print(son_puanlar)

ortalama = np.mean(son_puanlar)
maksimum = np.max(son_puanlar)
toplam  = np.sum(son_puanlar)
st_sapma = np.std(son_puanlar)

print("Puanların ortalaması: ",ortalama )
print("Puanların maksimumu: ",maksimum )
print("Puanların toplamı: ",toplam )
print("Puanların standart sapması: ",st_sapma )

gecenler = son_puanlar[son_puanlar > 50]
print("Sınavı geçenlerin notları= ", gecenler)

print("\n\n\n\n\n")



```
